# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading From Source

```python
postlinks_schema= StructType([
  StructField("_Id", IntegerType(), True),
  StructField("_PostId", IntegerType(), True),
  StructField("_RelatedPostId", IntegerType(), True),
  StructField("_LinkTypeId", IntegerType(), True),
  StructField("_CreationDate", StringType(), True)
  ])
```

```python
df= spark.read.format("xml").schema(postlinks_schema).option("rowTag","row").load("/Volumes/social_media/bronze/socialmedia_dataset/PostLinks.xml")
```

# Writing To Bronze

```python
df.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.post_links")
```

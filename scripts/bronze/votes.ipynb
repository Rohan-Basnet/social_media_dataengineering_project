# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading From Source

```python
vote_schema= StructType([
    StructField('_Id', IntegerType(), True),
    StructField('_PostId', IntegerType(), True),
    StructField('_VoteTypeId', IntegerType(), True),
    StructField('_CreationDate', StringType(), True)
])
```

```python
df= spark.read.format("xml").schema(vote_schema).option("rowTag", "row").load("/Volumes/social_media/bronze/socialmedia_dataset/Votes.xml")

```

```python
df_clean= df.select(
    df["_Id"].alias("id"),
    df["_PostId"].alias("post_id"),
    df["_VoteTypeId"].alias("vote_type_id"),
    df["_CreationDate"].alias("creation_date")
)
```

# Writing To Bronze

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.votes")
```

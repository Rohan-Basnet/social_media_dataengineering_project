# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading From Source

```python
tag_schema= StructType([
    StructField("_Id", IntegerType(), True),
    StructField("_TagName", StringType(), True),
    StructField("_Count", IntegerType(), True),
    StructField("_ExcerptPostId", IntegerType(), True),
    StructField("_WikiPostId", IntegerType(), True)
])
```

```python
df= spark.read.format("xml").schema(tag_schema).option("rowTag", "row").load("/Volumes/social_media/bronze/socialmedia_dataset/Tags.xml")
```

```python
df_clean= df.select([
    df["_Id"].alias("id"),
    df["_TagName"].alias("tag_name"),
    df["_Count"].alias("count"),
    df["_ExcerptPostId"].alias("excerpt_post_id"),
    df["_WikiPostId"].alias("wiki_post_id")
])
```

# Writing To Bronze

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.tags")
```

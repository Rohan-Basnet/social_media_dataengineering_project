# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading From Source

```python
posthistory_schema= StructType([
StructField("_Id",LongType(),True),
StructField("_PostHistoryTypeId",LongType(),True),
StructField("_PostId",LongType(),True),
StructField("_RevisionGUID",StringType(),True),
StructField("_Text",StringType(),True),
StructField("_CreationDate",TimestampType(),True),
StructField("_UserId",LongType(),True),
StructField("_UserDisplayName",StringType(),True),
StructField("_Comment",StringType(),True),
StructField("_ContentLicense",StringType(),True)
])
```

```python
df= spark.read.format("xml").option("rowTag","row").schema(posthistory_schema).load("/Volumes/social_media/bronze/socialmedia_dataset/PostHistory_*.xml")
```

```python
df_clean= df.select(
    df["_Id"].alias("id"),
    df["_PostHistoryTypeId"].alias("post_history_type_id"),
    df["_PostId"].alias("post_id"),
    df["_RevisionGUID"].alias("revision_guid"),
    df["_Text"].alias("text"),
    df["_CreationDate"].alias("creation_date"),
    df["_UserId"].alias("user_id"),
    df["_ContentLicense"].alias("content_license")
)
```

# Writing To Bronze

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.post_history")
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
comment_schema = StructType([
    StructField("_Id", IntegerType(), True),
    StructField("_PostId", IntegerType(), True),
    StructField("_Score", IntegerType(), True),
    StructField("_Text", StringType(), True),
    StructField("_CreationDate", StringType(), True),
    StructField("_UserId", IntegerType(), True)
])
```

```python
df= spark.read\
    .format("xml")\
    .option("rowTag","row")\
    .schema(comment_schema)\
    .load("/Volumes/social_media/bronze/socialmedia_dataset/Comments.xml")
```

```python
df_clean= df.select(
    df["_Id"].alias("id"),
    df["_PostId"].alias("post_id"),
    df["_Score"].alias("score"),
    df["_Text"].alias("text"),
    df["_CreationDate"].alias("creation_date"),
    df["_UserId"].alias("user_id")
)
```

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.comments")
```

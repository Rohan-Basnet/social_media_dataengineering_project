# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading from Source

```python
badge_schema= StructType([
    StructField('_Id', IntegerType(), True),
    StructField('_UserId', IntegerType(), True),
    StructField('_Name', StringType(), True),
    StructField('_Date', StringType(), True),
    StructField('_Class', IntegerType(), True),
    StructField('_TagBased', StringType(), True)
])
```

```python
df= spark.read.format("xml").option("rowTag","row").schema(badge_schema).load("/Volumes/social_media/bronze/socialmedia_dataset/Badges.xml")
```

```python
df_clean= df.select(
    df['_Id'].alias('id'),
    df['_UserId'].alias('user_id'),
    df['_Name'].alias('name'),
    df['_Date'].alias('date'),
    df['_Class'].alias('class'),
    df['_TagBased'].alias('tag_based')
)
```

# Writing To Bronze

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.badges")
```

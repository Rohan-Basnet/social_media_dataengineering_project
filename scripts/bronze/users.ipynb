# Initialization

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# Reading From Source

```python
user_schema= StructType([
    StructField('_Id', IntegerType(), True),
    StructField('_Reputation', IntegerType(), True),
    StructField('_CreationDate', TimestampType(), True),
    StructField('_DisplayName', StringType(), True),
    StructField('_LastAccessDate', TimestampType(), True),
    StructField('_UpVotes', IntegerType(), True),
    StructField('_DownVotes', IntegerType(), True),
    StructField('_Views', IntegerType(), True),
    StructField('_AccountId', IntegerType(), True)
])
```

```python
df= spark.read.format("xml").option("rowTag","row").schema(user_schema).load("/Volumes/social_media/bronze/socialmedia_dataset/Users.xml")
```

```python
df_clean= df.select(
    df['_Id'].alias('id'),
    df['_DisplayName'].alias('display_name'),
    df['_Reputation'].alias('reputation'),
    df['_CreationDate'].alias('creation_date'),
    df['_LastAccessDate'].alias('last_access_date'),
    df['_UpVotes'].alias('up_votes'),
    df['_DownVotes'].alias('down_votes'),
    df['_Views'].alias('views'),
    df['_AccountId'].alias('account_id')
)
```

# Writing To Bronze

```python
df_clean.write.mode("overwrite").format("delta").saveAsTable("social_media.bronze.users")
```
